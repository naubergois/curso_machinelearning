```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import numpy as np
```

```python
# 1. Number of times pregnant
# 2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
# 3. Diastolic blood pressure (mm Hg)
# 4. Triceps skin fold thickness (mm)
# 5. 2-Hour serum insulin (mu U/ml)
# 6. Body mass index (weight in kg/(height in m)^2)
# 7. Diabetes pedigree function
# 8. Age (years)
# 9. Class variable (0 or 1)
```

```python
diabetes_india=pd.read_csv('pima-indians-diabetes.csv',header=None)
```

```python
diabetes_india.head()
```

```python
clf=KNeighborsClassifier(n_neighbors=8)
```

```python
X=diabetes_india.drop(8,axis=1)
```

```python
y=diabetes_india[8]
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

precisão (também chamada de valor preditivo positivo) é a fração de instâncias recuperadas que são relevantes, enquanto revocação (também conhecida como sensibilidade) é a fração de instâncias relevantes que são recuperadas.
