```python
from sklearn.linear_model import Ridge
from sklearn.datasets import load_boston
import pandas as pd
```

```python
boston=load_boston()
```

```python
type(boston)
```

```python
print(boston.keys())
```

```python
print(boston['DESCR'])
```

```python
X=boston.data
```

```python
y=boston.target
```

```python
X=pd.DataFrame(X)
```

```python
X.head()
```

```python
type(X)
```

```python
X.columns=boston.feature_names
```

```python
X
```

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
scaler=MaxAbsScaler()
```

```python
scaler.fit(X)
```

```python
X_1=scaler.transform(X)
```

```python
X_1
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_1,y,test_size=0.2)
```

```python
for i in range(1,10):
    clf=KNeighborsRegressor(n_neighbors=i)
    clf.fit(X_train,y_train)
    print("Quantidade de vizinhos "+str(i))
    print(clf.score(X_test,y_test))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_2=scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_2,y)
```
